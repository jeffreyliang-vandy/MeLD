```python
import h5py
import pickle as pkl
import torch
import numpy as np

h5_path = "./data/dataset/ea_hiv/preprocessed/merged/train/data.h5"
pickle_path = "./data/dataset/ea_hiv/preprocessed/merged/train/parser.pkl"
def to_numpy(tensor):
    """Safely converts PyTorch tensors to NumPy arrays."""
    if isinstance(tensor, torch.Tensor):
        return tensor.cpu().numpy()
    return np.array(tensor)

print("2. Writing to HDF5 format...")
with h5py.File(h5_path, 'w') as h5f:
    # Save the massive arrays as individual datasets
    # chunks=True optimizes the file for lazy loading and slicing later
    h5f.create_dataset('processed_data', data=to_numpy(processed_data), chunks=True)
    h5f.create_dataset('time_info', data=to_numpy(time_info), chunks=True)
    h5f.create_dataset('missing', data=to_numpy(missing), chunks=True)
    h5f.create_dataset('masking', data=to_numpy(masking), chunks=True)

    # Save the datatype_info dictionary as HDF5 attributes (metadata)
    print("3. Saving datatype_info metadata...")
    for key, value in data_info.items():
        h5f.attrs[key] = value

print("3. Saving data parser")
with open(pickle_path,"wb") as f:
    pkl.dump(parser,f)

print(f"✅ Success! Your lazy-loadable file is saved at: {h5_path}")
```
